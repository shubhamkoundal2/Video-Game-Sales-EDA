# **About Dataset**
This dataset contains a list of video games with sales greater than 100,000 copies. It was generated by a scrape of **vgchartz.com**.

**Fields include**

Rank - Ranking of overall sales

Name - The games name

Platform - Platform of the games release (i.e. PC,PS4, etc.)

Year - Year of the game's release

Genre - Genre of the game

Publisher - Publisher of the game

NA_Sales - Sales in North America (in millions)

EU_Sales - Sales in Europe (in millions)

JP_Sales - Sales in Japan (in millions)

Other_Sales - Sales in the rest of the world (in millions)

Global_Sales - Total worldwide sales.

# **Exploratory Data Analysis & Data Visualization**

# **1. What is the Sales Trend by year?**

```python
sum_by_Year= df[['Year','NA','EU','JP','Global','Other']].groupby(['Year']).sum().reset_index()
```

```python
sum_by_Year
```

```python
fig = plt.figure(num=1, figsize =(12, 8))
plt.title('Sales Trends by Year', fontweight ='bold', fontsize = 20)
plt.xlabel('Year', fontweight ='bold', fontsize = 15)
plt.ylabel('Sales', fontweight ='bold', fontsize = 15)
for col in sum_by_Year:
    if not col == 'Year':
        plt.plot(sum_by_Year['Year'],sum_by_Year[col], label=col)
plt.legend()
plt.show()
```

# 2. What is the Genres Distribution of Percentage?

```python
Genres= df['Genre'].unique()
Genres
```

```python
genre_percentage = df['Genre'].value_counts(normalize=True) * 100
```

```python
genre_percentage
```

```python
plt.figure(figsize=(8, 6))
plt.pie(genre_percentage, labels=genre_percentage.index, autopct='%1.1f%%')
plt.title('Genre Distribution Percentage')
plt.axis('equal')
plt.show
```

# 3.What are the top 10 publishers of vedio games industry ?

```python
value_counts_publisher = df['Publisher'].value_counts()
top_10_publishers = value_counts_publisher.head(10)
top_10_publishers
```

```python
plt.figure(figsize=(12, 4))
ax = sns.barplot(x=top_10_publishers.values, y=top_10_publishers.index, palette='hls')
ax.set(xlabel='Count', ylabel='Publisher')
ax.set_title('Top 10 Publishers')
for i, v in enumerate(top_10_publishers.values):
    ax.text(v + 3, i, str(v), color='black', fontweight='bold')
plt.tight_layout()
plt.show()
```

# 4.Which Genres have the most sales in each territory.

```python

# Define the color palette
color = {
    'NA': ['#3333ff'],
    'EU': ['#1affa3'],
    'JP': ['#ffff33'],
    'Other': ['#ff3333'],
    'Global': ['#ffad33']
}

# Create a DataFrame from the color dictionary
color_df = pd.DataFrame(color)

# Iterate over the regions and plot the bar graph
for region in ['NA', 'EU', 'JP', 'Other', 'Global']:
    sales = df[['Genre', region]].groupby('Genre').sum().reset_index()
    sales = sales.rename(columns={region: 'Total_Sales'}).sort_values(by='Total_Sales', ascending=False)
    
    fig = plt.figure(num=1, figsize=(15, 5))
    plt.title('Total Sales in ' + region, fontweight='bold', fontsize=20)
    plt.xlabel('Genres', fontweight='bold', fontsize=15)
    plt.ylabel('Sales', fontweight='bold', fontsize=15)
    plt.bar(sales['Genre'], sales['Total_Sales'], color=color_df[region])
    plt.show()

```
